# 3.該怎麼算心跳速率呀...?




*   換成自己的心電訊號時，記得重新觀察上面的波形來調整 Distance(X軸) 和 Height(Y軸)，來取得最佳的Find_Peak表現唷！
*   小技巧:Y軸(height)-肉眼觀察最小的Peak值在Y軸上的值約略是落在多少呢?





```python
#1.執行程式
#2.觀察段落程式的說明
#3.往下查看這組訊號的波形圖



#偵測每一下心跳的波峰 

#find_peaks是一個偵測波形波峰的公式
#它有兩個參數，Height可以設定「至少Y軸的值是多少以上才能被算是波峰」、Distance可以設定「波峰和波峰之間至少要間隔多少」。
#實作前要先觀察上段程式碼秀出的波形唷！範例心電訊號和你自己記錄的心電訊號可是會有不同的height和distance範圍的呢！

height=90 
distance=250 #每格是0.002秒的話...distance應該設多少呢？
peak_list = find_peaks(data2, height=height, distance=distance)[0] 

#稍微設定一下畫布資訊！
fig = go.Figure()
fig.add_trace(go.Scatter(
    y=data2,
    name='心電訊號'
))

#把偵測到的波峰也標記在畫布上吧！
fig.add_trace(go.Scatter(
    x=peak_list,
    y=[data2[j] for j in peak_list],
    mode='markers',
    marker=dict(
        size=8,
        color='red',
        symbol='cross'
    ),
    name='分析之Peak位置'
))
fig.update_layout(
    title="範例心電訊號與波峰偵測", #換成自己的訊號時，可以在這行將title修改成自己的名字唷！
    font=dict(
        family="Courier New, monospace",
        size=20,
        color="RebeccaPurple"
    )
)
#秀出你的成果吧！
fig.show()
```

# 4.可以計算心跳速率了！

```python
#1.執行程式
#2.觀察段落程式的說明
#3.往下查看計算結果


print('每個心跳波峰是在資料位置',peak_list)

peak_peak_list = (np.diff(peak_list)) #兩兩波峰相減計算波峰之間的距離
print('心跳波峰之間分別距離幾個資料單位',peak_peak_list)

peak_peak_list2 = peak_peak_list*0.002 #每個資料單位之間的距離是0.002秒
print('每個心跳時間是',peak_peak_list2)

peak_peak_time = np.mean(peak_peak_list2) #把每個心跳時間加總取平均吧
print('每個心跳的時間平均間隔',peak_peak_time)

heart_rate = (60/peak_peak_time) #60秒內心跳可以跳幾下呢
print('心跳速率為',heart_rate, '下/分鐘')
```

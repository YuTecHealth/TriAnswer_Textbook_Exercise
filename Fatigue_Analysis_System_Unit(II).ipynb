# 3. How do we measure the heart rate...?




*   When switching to your own ECG signal, remember to use the figure above to adjust the distance(x-axis) and height(y-axis) to obtain the best find_peaks performance!
*   Tip: y-axis(height)-What is the approximate value of the smallest peak on the y-axis?





```python
#1.Execute Program
#2.Read the program's instruction
#3.Observe the figure below



#Observe the peak of each heart beat

#find_peaks is peak detecting formula
#It has two parameters, height can be set as [the minimum y level required to count as a wave], distance can be set as [the minimum distance between each wave]
#Observe the figure in the previous section first! The sample ECG signal may have different height and distance range from the ECG signal you recorded!

height=90
distance=250 #If each grid is 0.002 second...how much should the distance be?
peak_list = find_peaks(data2, height=height, distance=distance)[0]

#Config the figure setting!
fig = go.Figure()
fig.add_trace(go.Scatter(
    y=data2,
    name='ECG Signal'
))

#Record the detected peaks on the figure!
fig.add_trace(go.Scatter(
    x=peak_list,
    y=[data2[j] for j in peak_list],
    mode='markers',
    marker=dict(
        size=8,
        color='red',
        symbol='cross'
    ),
    name='Detected Peak Position'
))
fig.update_layout(
    title="Sample ECG Signal and Peak Detection", #When switching to your own signal, you can change the title on this row to any name you like!
    font=dict(
        family="Courier New, monospace",
        size=20,
        color="RebeccaPurple"
    )
)
#Display your result!
fig.show()
```

# 4. Calculate the heart rate!

```python
#1.Execute Program
#2.Read the program's instruction
#3.Observe the calculation result below


print('Position of each ECG peak',peak_list)

peak_peak_list = (np.diff(peak_list)) #Get the difference between the peak to get the unit distance
print('Unit distance between each ECG peak',peak_peak_list)

peak_peak_list2 = peak_peak_list*0.002 #Each unit distance is 0.002 second
print('Time between each heartbeat is',peak_peak_list2)

peak_peak_time = np.mean(peak_peak_list2) #Sum all the heart beat time and get the average
print('Average time between heartbeats',peak_peak_time)

heart_rate = (60/peak_peak_time) #How much time does the heart beat in 60 seconds
print('Heart rate is',heart_rate, 'beats per minute')
```
